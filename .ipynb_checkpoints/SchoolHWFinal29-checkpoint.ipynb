```python
import pandas as pd
```

```python
school_data_to_load =  "C:/Users/User/School.csv"
student_data_to_load = "C:/Users/User/Student.csv"
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
total_numbers_schools= school_data_complete["School ID"].nunique()
Total_students= school_data_complete["Student ID"].count()
total_budget= school_data_complete["budget"].unique().sum()
Average_Math_score= (school_data_complete["math_score"]).mean()
Average_Reading_score= (school_data_complete["reading_score"]).mean()
Total_Passing_rate= (Average_Math_score+Average_Reading_score)/2
Passing_Math=(school_data_complete['math_score'].loc[school_data_complete["math_score"]>=70].count()/Total_students)*100
Passing_Reading= (school_data_complete["reading_score"].loc[school_data_complete["reading_score"]>=70].count()/Total_students)*100
```

```python
Distirict_Summary = pd.DataFrame({"Total Schools":[total_numbers_schools],
                           "Total Students":[Total_students], 
                           "Total Budget": [total_budget], 
                           "Average Math Score": [Average_Math_score] ,
                                 "Average Reading Score":[Average_Reading_score],
                                 "%Passing Math":[Passing_Math],
                                  "%Passing Reading":[Passing_Reading],
                                  "%Overall Passing Rate":[Total_Passing_rate]})
Distirict_Summary.style.format({'Total Budget':"${:,.2f}"})
```

```python
grade9 = school_data_complete[school_data_complete["grade"] == "9th"]
grade10 = school_data_complete[school_data_complete["grade"] == "10th"]
grade11 = school_data_complete[school_data_complete["grade"] == "11th"]
grade12 = school_data_complete[school_data_complete["grade"] == "12th"]
group9 = grade9.groupby("school_name")
group10 = grade10.groupby("school_name")
group11 = grade11.groupby("school_name")
group12 = grade12.groupby("school_name")
m1 = group9['math_score'].mean()
m2 = group10['math_score'].mean()
m3 = group11['math_score'].mean()
m4 = group12['math_score'].mean()
r1 = group9['reading_score'].mean()
r2 = group10['reading_score'].mean()
r3 = group11['reading_score'].mean()
r4 = group12['reading_score'].mean()
```

```python
bins = [0, 585, 615, 645, 675]
names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
school_data_complete['spending_bins2'] = pd.cut(school_data_complete['size'], size_bins, labels = group_names )

by_spending = school_data_complete.groupby('spending_bins2')

avg_math = by_spending['math_score'].mean()
avg_reading = by_spending['reading_score'].mean()


passread = school_data_complete[school_data_complete["reading_score"]>= 70].groupby('spending_bins2')["Student ID"].count()/by_spending["Student ID"].count()
passmath = school_data_complete[school_data_complete["math_score"]>= 70].groupby('spending_bins2')["Student ID"].count()/by_spending["Student ID"].count()

SummaryTable = pd.DataFrame({"Average Math Score" : avg_math, "Average Reading Score" : avg_reading, "%Passing Math" : passmath, "% Passing Reading":passread})

SummaryTable["%Overall Passing Rate"] = (SummaryTable["%Passing Math"] + SummaryTable["% Passing Reading"])/2

SummaryTable.style.format({"Average Math Score": '{:.1f}',
                           "Average Reading Score" : '{:.1f}',
                           '%Passing Math': '{:.1%}',
                           '% Passing Reading':'{:.1%}',
                           '%Overall Passing Rate': '{:.1%}'
                          
                          })
```
